# E-commerce Sales Analysis using GroupBy and Aggregations

This comprehensive notebook demonstrates advanced Pandas grouping and aggregation techniques for sales analysis. We'll analyze synthetic e-commerce data to identify top performers, regional trends, and product category performance.

## Dataset Creation

Generate a synthetic sales dataset with 200 transactions containing information about regions, product categories, salespersons, and sales amounts.

```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print("=== DATASET PREVIEW ===\n")
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
print(f"\nColumn data types:\n{df.dtypes}")
```

## Task 1: Basic Grouping and Single Aggregations (35 points)

Perform fundamental grouping operations with single aggregation functions to understand:
- Regional sales performance
- Transaction distribution across categories
- Average sales by salesperson

### 1.1: Total Sales Amount by Region
Group by region and sum the sales amounts to identify which geographical area generates the most revenue.

```python
# Calculate total sales amount for each region
regional_sales = df.groupby('region')['sales_amount'].sum().reset_index()
regional_sales.columns = ['region', 'total_sales']

print("=== SALES BY REGION ===\n")
print(regional_sales)

# Sort to find top-performing region
regional_sales_sorted = regional_sales.sort_values('total_sales', ascending=False)
print("\n--- Regional Sales Ranked (Highest to Lowest) ---")
print(regional_sales_sorted)
print(f"\nTop-Performing Region: {regional_sales_sorted.iloc[0]['region']} with ${regional_sales_sorted.iloc[0]['total_sales']:,.2f}")
```

### 1.2: Transaction Count by Product Category
Count the number of transactions for each product category to understand which products are most frequently sold.

```python
# Count the number of transactions for each product category
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()
category_transactions.columns = ['category', 'transaction_count']

print("=== TRANSACTIONS BY CATEGORY ===\n")
print(category_transactions)

# Sort to find most popular category
category_sorted = category_transactions.sort_values('transaction_count', ascending=False)
print("\n--- Categories Ranked by Transaction Count ---")
print(category_sorted)
print(f"\nMost Transacted Category: {category_sorted.iloc[0]['category']} with {category_sorted.iloc[0]['transaction_count']} transactions")
```

### 1.3: Average Sales per Salesperson
Calculate the mean sales amount for each salesperson to measure their average deal size.

```python
# Calculate the average sales amount per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()
salesperson_avg.columns = ['salesperson', 'average_sales']

print("=== AVERAGE SALES BY SALESPERSON ===\n")
print(salesperson_avg)

# Sort to find highest average
salesperson_sorted = salesperson_avg.sort_values('average_sales', ascending=False)
print("\n--- Salespersons Ranked by Average Sale Size ---")
print(salesperson_sorted)
print(f"\nHighest Average Sales: {salesperson_sorted.iloc[0]['salesperson']} with ${salesperson_sorted.iloc[0]['average_sales']:,.2f} average sale")
```

### 2.1: Sales by Region and Category Combination
Analyze total sales for each region-category pair to identify which combinations perform best.

```python
# Group by both region AND category to calculate total sales for each combination
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
region_category_sales.columns = ['region', 'category', 'total_sales']

print("=== SALES BY REGION AND CATEGORY ===\n")
print(region_category_sales)

# Find the best performing region-category combination
best_combo = region_category_sales.loc[region_category_sales['total_sales'].idxmax()]
print(f"\n--- Top Region-Category Combination ---")
print(f"{best_combo['region']} - {best_combo['category']}: ${best_combo['total_sales']:,.2f}")

# Show pivot table view for better visualization
print("\n--- Pivot Table View (Region vs Category) ---")
pivot_sales = region_category_sales.pivot(index='region', columns='category', values='total_sales')
print(pivot_sales.fillna(0))
```

### 2.2: Comprehensive Salesperson Metrics
Calculate total sales, average transaction size, and transaction count for each salesperson using multiple aggregations.

```python
# Calculate multiple metrics for each salesperson simultaneously
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
salesperson_metrics.columns = ['salesperson', 'total_sales', 'average_sales', 'transaction_count']

print("=== COMPREHENSIVE SALESPERSON METRICS ===\n")
print(salesperson_metrics)

# Sort by total sales to identify top performer
salesperson_ranked = salesperson_metrics.sort_values('total_sales', ascending=False)
print("\n--- Salespersons Ranked by Total Sales ---")
print(salesperson_ranked)
print(f"\n✓ Top Salesperson: {salesperson_ranked.iloc[0]['salesperson']}")
print(f"  - Total Sales: ${salesperson_ranked.iloc[0]['total_sales']:,.2f}")
print(f"  - Avg Transaction: ${salesperson_ranked.iloc[0]['average_sales']:,.2f}")
print(f"  - Transactions: {int(salesperson_ranked.iloc[0]['transaction_count'])}")
```

### 2.3: Identify Category with Maximum Revenue
Use idxmax() to programmatically find which product category generates the highest total revenue.

```python
# Calculate total sales by category
category_sales = df.groupby('category')['sales_amount'].sum().reset_index()
category_sales.columns = ['category', 'total_sales']

print("=== SALES BY CATEGORY (SORTED) ===\n")
print(category_sales.sort_values('total_sales', ascending=False))

# Use idxmax() to find the category with maximum revenue
max_revenue_idx = category_sales['total_sales'].idxmax()
max_revenue_category = category_sales.loc[max_revenue_idx]

print(f"\n--- Category with Maximum Revenue ---")
print(f"Category: {max_revenue_category['category']}")
print(f"Total Revenue: ${max_revenue_category['total_sales']:,.2f}")

# Show percentage contribution of each category
total_sales = category_sales['total_sales'].sum()
category_sales['percentage'] = (category_sales['total_sales'] / total_sales * 100).round(2)
print("\n--- Revenue Distribution by Category ---")
print(category_sales.sort_values('total_sales', ascending=False))
```

## Task 3: Custom Aggregation and Complete Sales Report (30 points)

Create advanced aggregations using custom functions and generate a comprehensive multi-level summary report.

### 3.1: Define and Apply Custom Aggregation Function
Create a custom function to calculate the sales range (max - min) for each grouping.

```python
# Define a custom aggregation function that calculates sales range
def sales_range(x):
    """Calculate the range (max - min) of sales amounts"""
    return x.max() - x.min()

# Apply custom function along with standard aggregations to analyze sales by region
regional_analysis = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
regional_analysis.columns = ['region', 'total_sales', 'average_sales', 'min_sale', 'max_sale', 'sales_range']

print("=== COMPREHENSIVE REGIONAL ANALYSIS ===\n")
print(regional_analysis)

print("\n--- Detailed Explanation ---")
for idx, row in regional_analysis.iterrows():
    print(f"\n{row['region']}:")
    print(f"  Total Sales: ${row['total_sales']:,.2f}")
    print(f"  Average Sale: ${row['average_sales']:,.2f}")
    print(f"  Min Sale: ${row['min_sale']:,.2f}")
    print(f"  Max Sale: ${row['max_sale']:,.2f}")
    print(f"  Range: ${row['sales_range']:,.2f}")
```

### 3.2: Multi-Column Dictionary-Based Aggregation
Create a final summary report using dictionary syntax to apply different aggregations to different columns simultaneously.

```python
# Create final summary report with dictionary-based aggregation
summary_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("=== FINAL SUMMARY REPORT BY REGION ===\n")
print(summary_report)

# Rename columns for clarity
summary_report.columns = ['region', 'total_sales', 'average_transaction_value', 'transaction_count']

print("\n--- Cleaned Summary Report ---")
print(summary_report)

print("\n--- Key Insights ---")
summary_report_sorted = summary_report.sort_values('total_sales', ascending=False)
for idx, row in summary_report_sorted.iterrows():
    print(f"\n{row['region']} Region:")
    print(f"  Total Sales: ${row['total_sales']:,.2f}")
    print(f"  Avg Transaction Value: ${row['average_transaction_value']:,.2f}")
    print(f"  Number of Transactions: {int(row['transaction_count'])}")
```

### 3.3: Explanation of Multi-Level Column Structure
Explore the multi-level (hierarchical) column structure that results from dictionary-based aggregation.

```python
# Re-create the aggregation to show multi-level structure
summary_with_multilevel = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
})

print("=== MULTI-LEVEL COLUMN STRUCTURE ===\n")
print("DataFrame with Multi-Level columns:")
print(summary_with_multilevel)

print("\n--- Column Structure Information ---")
print(f"Column levels: {summary_with_multilevel.columns.nlevels}")
print(f"Level 0 (columns): {list(summary_with_multilevel.columns.get_level_values(0).unique())}")
print(f"Level 1 (aggregations): {list(summary_with_multilevel.columns.get_level_values(1).unique())}")

print("\n--- Accessing Multi-Level Columns ---")
print("Access specific column: summary_with_multilevel[('sales_amount', 'sum')]")
print(summary_with_multilevel[('sales_amount', 'sum')])

print("\n--- Advantages of Multi-Level Columns ---")
print("✓ Preserves relationship between column and aggregation function")
print("✓ Easily access specific aggregations: df[(column, agg_func)]")
print("✓ Better for large datasets with many aggregations")
print("✓ Clearer structure when exporting to Excel or other formats")

print("\n--- Difference from Single Aggregations ---")
print("Single aggregation (e.g., df.groupby('region')['sales_amount'].sum()):")
print("  - Returns Series with region as index")
print("  - Simple structure, flat column names")
print("  - Better for quick, focused analysis")
print("\nMultiple aggregations with dictionary:")
print("  - Returns DataFrame with multi-level columns")
print("  - Complex structure, hierarchical column names")
print("  - Better for comprehensive reports across multiple columns")
```
